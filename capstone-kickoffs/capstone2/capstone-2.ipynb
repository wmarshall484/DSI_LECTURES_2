## Capstone 2 project proposal

We'll be working on the same project during Capstone 2 and 3. The proposal for this project will be more detailed and more specific than Capstone 1. Specifically, whereas the last project was focused on exploration, this capstone needs to **solve a specific problem**. The problem could be from the world of:
* Business
* Nonprofit organizations
* Government
* Academic research
* Open source tools

Often we say that the project should "address a business case"; here "business case" means "specific real-world problem" not necessarily limited to the functioning of a for-profit business.

### Generating project proposals

Here's a strategy for generating a project proposal

1. Think of a frustrating thing; some kind of struggle or pain-point.
2. Think of a tool that would address or alleviate that struggle.
3. See if you can express that tool in terms of one of the aspects of "data science"
   * Supervised learning
      * Classification
      * Regression
   * Unsupervised learning
      * Clustering
      * Topic modeling
   * Recommender system
   * Hypothesis testing
   * Network analysis

### Examples

* Taxi driver
  * Problem: Taxi driver doesn't get any riders.
  * Tool: App says where to go where there will people will be hailing cabs.
  * Data science: **Supervised regression**: Given a location and time, predicts how long until a rider hails a cab.
  
* Online store
  * Problem: Don't know how to organize inventory onto website
  * Tool: Upload images and descriptions of products, automatically organizes it
  * Data science: **Heirarchical Clustering** vectorize images and decscriptions and perform heirarchical clustering to give a "directory" structure of inventory
  
* Political campaigns
  * Problem: if you're competing in many districts, where should you focus your campaign?
  * Tool: A system that analyzes polling data to give a confidence interval of the vote outcome
  * Data science: **Hypothesis testing** find a probability distribution of a bernoulli distribution represnting the vote, given a number of samples from that distribution.
  
* Reading
  * Problem: I wish I could read a very specific book that doesn't exist
  * Tool: Select some books, and then find books that "combine" them.
  * Data science: **Topic Modeling** find latent topics of input books, combine their latent topic vectors, and find books that are similar to the combined vector.

Make a list of *more than ten* problems, and then use this process to see if you can generate a data science project. Not every idea will yield a project. That's fine. A few will.

## Deliverable: Minimum Viable Product

We only have about four class-days to work on this project. As a result, the scope will be very limited.

By the end of next week, you should have a "minimum viable product" ("MVP") implementation of your idea. The MVP will consist of working code that:

* Loads your data
* Vectorizes it into features
  * If supervised, also finds the target
* Fits a model
* If supervised:
  * Predict on the test set
  * Calculate a test metric
* If unsupervised
  * Transform the dataset
  * Show the transformation in a straightforward way
    * In "online store" example, show the inventory directory structure.
    * In the "book combiner" example, show the search results for combining two books.
    
That's all. You don't have to make a website or a fancy presentation, or chase down every improvement you think while you're working. That's what Capstone 3 is for.
