## SQL

#### Loading a database

* in your terminal, navigate to the folder containing the databasename.sql file
* `psql` to enter the postgres terminal
  * `CREATE DATABASE databasename;` where `databasename` is the name of the database
  * `\q` to exit the postgres terminal
* `psql databasename < databasename.sql` to load the database
* `psql databasename` to log in to the database and start SQLing!!

# Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv('pandas/data/hospital-costs.csv')
```

```python
df.info()
```

```python
df.head()
```

Add a new column 'Total Cost'

```python
df['Total_Cost'] = df['Discharges'] * df['Mean Cost']
```

```python
df.head()
```

Group by facility name and get the total cost for each hospital

```python
df.groupby('Facility Name').sum()
```

```python
df.groupby('Facility Name').sum()['Facility Name']
```

```python
df.groupby('Facility Name').sum().reset_index()
```

```python
df.groupby('Facility Name').sum()['Total_Cost'].head()
```

```python
df.groupby('Facility Name').sum().iloc[:,-1].head()
```

```python
df[['Facility Name', 'Total_Cost']].groupby('Facility Name').sum().reset_index()
```

```python
df[['Facility Name', 'Total_Cost']].groupby('Facility Name').sum().reset_index().iloc[0,1]
```
