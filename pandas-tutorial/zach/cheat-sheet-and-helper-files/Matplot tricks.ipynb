````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(3, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(3, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
plt.subplot(3, 1, 3)
plt.plot(x1, y1, color='green', linestyle='dashed', marker='o',
     markerfacecolor='blue', markersize=4)
````

## ```fig.subplot()``` is confusing as fudge, 
* Because the first two arguments for it are number_of_horizontal_charts, then, number_of_vertical_charts,  then, the position of the chart 
* The first two argumensts stay the same, 
* The third one chanages
* ```fig.subplot(1, 1, 1)```
* ```chart1 = fig.subplot(total_charts_down, total_charts_across , sub_chart_position)```

````python
fig = plt.figure()
total_charts_down = 2
total_charts_across = 3

chart1 = fig.add_subplot(total_charts_down, total_charts_across, 1, axisbg='red')
chart2 = fig.add_subplot(total_charts_down, total_charts_across, 2, axisbg='orange')
chart3 = fig.add_subplot(total_charts_down, total_charts_across, 3, axisbg='yellow')
chart4 = fig.add_subplot(total_charts_down, total_charts_across, 4, axisbg='green')
chart5 = fig.add_subplot(total_charts_down, total_charts_across, 5, axisbg='blue')
chart6 = fig.add_subplot(total_charts_down, total_charts_across, 6, axisbg='violet')

### THIS LOOKS LIKE A GOOD JOB FOR A LOOP
````
