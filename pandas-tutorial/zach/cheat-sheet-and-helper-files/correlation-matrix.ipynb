```python
dfsals = pd.read_csv('../data/baseball-csvs/Salaries.csv')
dfsals.head()

## I WANT TO KNOW IF THERE IS A CORRELATION BETWEEN 
## BATTING AVERAGE AND SALARY 
## JUST FOR THE YEAR 2013

## IF WE JUST WANT TO LOOK AT THE YEAR 2013,
## SOO... RETURN THE DF WHEN THE DF COLUMN YEARID COL IS 2013

sub_sals = dfsals[ dfsals['yearID'] == 2013 ]
sub_sals
```

```python
dfbatting = pd.read_csv('../data/baseball-csvs/Batting.csv')
sub_bat = dfbatting[ dfbatting['yearID'] == 2013 ]
```

```python
mergeddf = pd.merge(sub_bat, sub_sals, on='playerID')
```

```python
## SINCE THERE IS NOT BATTING AVERAGE, WHICH IS CRAY CRAY,
## WE HAVE TO CALUCLATE IT OURSELVES.  FIGURE IT OUT
mergeddf['BA'] = mergeddf['H'] / mergeddf['AB']
```

```python
ba_cor_sal = mergeddf[['salary', 'BA']].corr()
```
