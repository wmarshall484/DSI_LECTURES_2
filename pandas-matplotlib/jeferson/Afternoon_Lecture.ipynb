```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#sns.set_style("white")
```

```python
crimes = pd.read_csv('data/crime.csv')
crimes.head()
```

```python
print "For minimum crimes"
for i in range(1, 8):
    print crimes.columns[i], ':', crimes.iloc[np.argmin(crimes.iloc[:,i]), 0]
print
print "For maximum crimes"
for i in range(1, 8):
    print crimes.columns[i], ':', crimes.iloc[np.argmax(crimes.iloc[:,i]), 0]
```

```python
crimes2 = crimes.iloc[:,1:] # Excluding state names.
```

```python
f, ax = plt.subplots(figsize=(5,25))
for i in range(7):
    plt.subplot(7, 1, i+1)
    plt.title(crimes2.columns[i])
    plt.hist(crimes2.iloc[:,i])
plt.savefig('histograms_plt.png')
```

```python
f, ax = plt.subplots(figsize=(5,25))
for i in range(7):
    plt.subplot(7, 1, i+1)
    sns.distplot(crimes2.iloc[:,i])
plt.savefig('histograms_sns.png')
```

```python
sns.heatmap(crimes2.corr())
plt.savefig('heatmap2.png')
```

```python
plt.figure(figsize=(10,10))
sns.corrplot(crimes2)
plt.savefig('heatmap.png')
```

```python
sns.set_style("white")
sns.set_context("paper")
# f, ax = plt.subplots(figsize=(12,6))
sns.boxplot(crimes2)
sns.despine(left=True, bottom=True)
plt.savefig('boxplot.png')
```

```python
f, ax = plt.subplots(figsize=(12,6))
sns.violinplot(crimes2)
sns.despine(left=True, bottom=True)
# plt.savefig('violinplot.png')
```

```python
sns.violinplot(crimes2.iloc[:,:2])
sns.despine(left=True, bottom=True)
plt.savefig('violinplot2.png')
```
