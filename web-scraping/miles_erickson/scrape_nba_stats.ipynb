```python
def scrape_table(url, sel, delay=5):
    """Return a list of lists (rows) of data from a table.
    
    Arguments
    ---------
    url : str
        The URL of the site to scrape.
    sel : str
        The CSS selector of the table to scrape.
    """
    browser.get(url)
    time.sleep(delay)
    table = browser.find_element_by_css_selector(sel)
    table_rows = table.find_elements_by_css_selector('tr')
    data_rows = []
    for table_row in table_rows:
        th_elements = table_row.find_elements_by_css_selector('th')
        th_data = [th.text for th in th_elements]
        td_elements = table_row.find_elements_by_css_selector('td')
        td_data = [td.text for td in td_elements]
        data_rows.append(th_data + td_data)
    return data_rows
    #table_rows = table.text.split('\n')
    #return [row.split() for row in table_rows[1:]]
```

```python
def scrape_table_bs4(url, sel, delay=5):
    """Return a list of lists (rows) of data from a table.
    
    Arguments
    ---------
    url : str
        The URL of the site to scrape.
    sel : str
        The CSS selector of the table to scrape.
    """
    response = requests.get(url)
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    time.sleep(delay)
    table = soup.select_one(sel)
    table_rows = table.select('tr')
    data_rows = []
    for table_row in table_rows:
        th_elements = table_row.select('th')
        th_data = [th.text for th in th_elements]
        td_elements = table_row.select('td')
        td_data = [td.text for td in td_elements]
        data_rows.append(th_data + td_data)
    return data_rows
```
