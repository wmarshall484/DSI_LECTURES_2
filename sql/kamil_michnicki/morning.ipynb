### Why use a database management system?

Why use databases? You could store data in a file and have everyone share it. What are some problems with this? 

#### Concurrency and transactions:
<ul>
<li>Two people reading the same data is fine.</li>
<li>What if one person starts reading while another person reads. What version will they get? Will it be a consistent datafile?</li>
<li>What if one person is halfway done inserting a row while the other starts to read the row?</li>
</ul>

#### Fault tolerance
<ul>
<li>
What if the computer crashes or the connection is lost? Transactions protect the database.</li>
</ul>

#### Data independence
<ul>
<li>
What happens when the format of the data is changed within the text file? Do you now need to change all your code? How data is handled is abstracted from how data is managed.
</li>
</ul>

### What is a DBMS?  
<b>DBMS are systems that solve all of the problems that we had with storing data in a shared text file.</b> They allow us concurrently access and change data in a fault tolerant way which abstracts the way data is managed from how it is used in an application. This is all done for us under the hood, behind the scenes by the DBMS and quickly.

### What is a relational database?
<b>What is a relational database?</b> <b><i>Essentially tables with rows and named columns.</i></b> These columns are called fields. A set of fields which are associated with the same row of data is called a relation. Think of yelp. The restaurant id name and type of food are a relation. They are related to each other. Different tables are also related to each other. Restaurants are related to reviews which are related to users.

#### Create table 

CREATE TABLE restaurants 
(id SERIAL PRIMARY KEY, name VARCHAR(20),type VARCHAR(15));


    CREATE TABLE - does what it says.

    restaurants - the name of our table.

    id - the name of the first field(or column) of our table. 

    SERIAL - every field needs a datatype. This type automatically increments this column when a new row is inserted.

    PRIMARY KEY - Think of a dictionary. Each workd is a key. Keys are ways of looking entries in a table. A primary key makes sure every row is uniquely identified. If you try to insert a row with the same id as a previous id, because it is a primary key, you will get an error. 

    name - the name of our second field(or column) of our table.

    VARCHAR or VARCHAR(20) - variable length string of characters. In this case, at most 20 characters long.  

    ; - every query ends with ;

There are many more datatypes:

What does the following do?

We can relate the tables with a WHERE clause.

SELECT * FROM restaurants, reviews 
WHERE restuarants.id=reviews.restaurant_id;

    restaurants.id=reviews.restaurant_id - Two tables may share the same field name so we have to specify which table the field comes from. We do this by writing tablename.fieldname.

Would be nice if they were ordered...

SELECT * FROM restaurants, reviews 
WHERE restuarants.id=reviews.restaurant_id
ORDER BY restaurants.id;


Equivalently, we can do a join.

SELECT * FROM restaurants 
JOIN reviews ON restaurants.id=reviews.restaurant_id;

Or

SELECT * FROM restaurants 
INNER JOIN reviews ON restaurants.id=reviews.restaurant_id;

Sometimes a user will have left a review and have not logged in. There will be no user id for that review. 

SELECT * FROM restaurants 
JOIN reviews ON restaurants.id=reviews.restaurant_id
LEFT JOIN users ON users.id=reviews.user_id;

We are getting a lot of fields now. We specify fields with table.field

SELECT restaurants.name, users.name, reviews.rating 
FROM restaurants 
JOIN reviews ON restaurants.id=reviews.restaurant_id
LEFT JOIN users ON users.id=reviews.user_id 
ORDER BY restaurants.id;

For LEFT JOIN, each row of the left table is joined to the right table and if there is no match then the right table returns NULL results for each field. RIGHT JOIN does the same but returns NULL results for the left table. 

SELECT restaurants.name, users.name, reviews.rating 
FROM users 
RIGHT JOIN reviews ON users.id=reviews.user_id
JOIN restaurants ON restaurants.id=reviews.restaurant_id
ORDER BY restaurants.id;

We can do both a LEFT JOIN and a RIGHT JOIN by using the FULL JOIN clauses 

```sql 
SELECT restaurants.name, users.name, reviews.rating 
FROM users 
FULL JOIN reviews ON users.id=reviews.user_id
JOIN restaurants ON restaurants.id=reviews.restaurant_id
ORDER BY restaurants.id;
```

#### Group by

Now let's count the number of reviews for each restaurant.

SELECT restaurants.name,count(reviews.restaurant_id) 
FROM restaurants 
JOIN reviews ON restaurants.id=reviews.restaurant_id 
GROUP BY restaurants.id;

It's not displaying restaurants with zero reviews. Fixed with a left join!

SELECT restaurants.name,count(reviews.restaurant_id) 
FROM restaurants 
LEFT JOIN reviews ON restaurants.id=reviews.restaurant_id 
GROUP BY restaurants.id;



#### Case

SELECT restaurants.name,

CASE WHEN count(reviews.restaurant_id)>0 
THEN 'Has reviews.'
ELSE 'Has no reviews.' END as has_reviews

FROM restaurants 
LEFT JOIN reviews ON restaurants.id=reviews.restaurant_id 
GROUP BY restaurants.id;


### Order of operations.

SELECT * FROM table JOIN table2 ON table.f=table2.g WHERE condition GROUP BY field ORDER BY field

FROM
JOIN
WHERE
GROUP BY
HAVING
SELECT
DISTINCT
ORDER BY

### Indexing: 

CREATE UNIQUE INDEX ON restaurants (id);

### Dumping a database:

pg_dump mydb > db.sql

### Moving data to a text file

COPY (SELECT * FROM restaurants WHERE type='Coffee and tea')
TO '/Users/kamilmichnicki/mylectures/sql/kamil_michnicki/cafes.copy' 
WITH DELIMITER ',';

Discussion

### Write simple queries on a single table including SELECT, FROM, WHERE, CASE clauses and aggregates.

### Write complex queries including JOINS and subqueries.

### Format a query to follow a standard style.

# Afternoon
<ul>
<li>##Connect to a database from within a python program.</li>
<li>Explain how indexing works in Postgres.</li>
<li>Create and dump tables.</li>
<li>Move data from SQL database to text file.</li>
<li>Discussion</li>
</ul>

It was a dark night and a super hero is on was on their computer. This super hero is personally responsible for putting many criminals behind bars at the Institute for Data Criminals, the IDC for short. Our hero promised that he would create a database for them. They want to manage the data about their inmates. Having a connection to this database was very important for the our hero. One never knows when one needs to do an investigation. But before our hero could do anything, our hero would need to connect to the database...

Our hero openned the terminal and sshed into the ICI server. They typed:

$ssh batman@idc.com

After entering the password, our hero could create the database by typing:

$createdb idc

And then connect to it by typing:

$psql idc

Our hero is now connected to the database. Now, our hero could have done this in one command:

$psql -h <host> -p <port> -U <username> -W <password> idc

If the database were on our heroes local machine, our hero could have typed:

$psql idc

Well anyway, back to the story. Now our hero is in an interactive terminal for posgresql. The first thing to do is to create a table.

$CREATE TABLE inmates (id SERIAL PRIMARY KEY, name VARCHAR(20))





the internet...

First he went to google and searched for recent video of a masked figure who had released a menacing video.
<a href='google.com'>Google</a>

Google's database churned as it pumped results out from it's many distributed servers. Much like the service men and women who protected Gotham city. Each working independently but still together.

<a href='youtube.com'>Youtube</a>

The threat was both vague but very real. Somewhere in the city, something bad would happen, unless, batman could figure out the mystery.


Batman went to the database for Gotham's institute for the criminally insane. 

He typed:

SELECT * FROM inmates;

and got all of the information about the inmates table.
 
id|name|status
1|Joker|inmate
2|Scar-face|inmate
...

Next he wanted to see who had commited which which crimes. So he typed:

SELECT * FROM crimes;

id|crime|inmate_id
1|pick-pocket|3
2|menacing threats|4
3|practical jokes|2
...

But then he discovered that he could not see the name of the person who had commited the crime, just their ids. He would need a way to join the tables together. After searching stackoverflow.com he found two different ways to do it.

SELECT * FROM inmates, crimes WHERE inmates.id=crimes.inmate_id;

SELECT * FROM inmates JOIN crimes on inmates.id=crimes.inmate_id;

They were similar but using the join seemed not to repeat the two fields that contained the inmate id. Batman did not like to waste space so he chose to use the JOIN. 

Meanwhile, Gotham's police force was hard at work. The plan was that the police force would map out any suspisious behavrior. They went to every neighborhood and asked around. each police officer would go to a different part of town and look for anything suspicious. They would come back to HQ and report.


The police force went on red alter. Batman was alerted to check the database.

SELECT * FROM inmates;
