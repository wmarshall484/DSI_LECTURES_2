# Normal Distribution

The *normal* or *gaussian* distribution has a hallowed place in statistical folklore, mostly due to the *central limit theorem* which we will study shortly.

## First, create code to calculate normal distributions.

If you are running the sample code yourself as you read this section, you need to import some dependencies to make the code run. Place the following snippet of code at the top of your Jupyter notebook or source code file:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from numpy import random as rnd
%matplotlib inline

plt.style.use('ggplot')
```

## PDF of the standard normal distribution

```python
mu = 0
sigma = 1
x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 1, figsize=(10, 5))
axs.plot(x, stats.norm.pdf(x, loc=mu, scale=sigma), linewidth=2)
axs.set_ylim(-0.05, 0.5)
axs.set_xlim(-4, 4)
axs.set_title("PDF from Normal(mu=0, sigma=1)")
```

Parameters of the Normal Distribution are μ and σ

+ The μ parameter is the mean.
+ σ is the standard deviation. σ2 is the variance. Both are measures of dispersion around the mean. The standard deviation is easier to interpret since it has the same measurement units of population that created the normal distribution.

Together, these parameters change the shape of the normal distribution probability density function (PDF) and cumulative desnity function (CDF). The height of y-axis is the proportion of the population at that value.

The CDF shows the proportion of the population that lies below that point in the population.

It is also useful to remember the 68-95-99.7 rules:

+ 68% of the population lies within +/- 1σ standard deviation of the the mean
+ 95% of the population lies within +/- 2σ standard deviations of the the mean μ
+ 99.7% of the population lies within +/- 3σ standard deviations of the the mean μ

What units are on the x-axis of the PDF and CDF? The x-axis is where the standard deviation and variance (σ and σ2) lie. If σ = 1, then the units on the x-axis are standard deviations.

### PMFs and PDFs

The smooth continuous curves that you see in this lesson are from probability density functions. The smooth curves come from continuous values. Probability mass functions (PMFs), in contrast, are made of counts of discrete values, or counts into bins of continuous values within certain ranges. They are represented as histograms which you will see below.

### CDFs

Cumulative distribution funcitons (CDFs) show *how much of the population lies at a value below the area on the curve*. As you go across the curve, notice how the value is always going up, and how the slope is steeper where the mean of the PDF lies. That is showing the accumulation of values, and that is what the CDF does.

## CDF of standard normal distribution

```python
mu = 0
sigma = 1
x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 1, figsize=(10, 5))
axs.plot(x, stats.norm.cdf(x, loc=mu, scale=sigma), linewidth=2)
# axs.set_ylim(-0.05, 0.5)
axs.set_xlim(-4, 4)
axs.set_ylim(-0.1, 1.1)
axs.set_title("CDF from Normal(mu=0, sigma=1)")
```

## Changing μ translates (shifts) the distribution function to the right and left

```python
mus = [-2, 0, 2]
sigma = 1.0

x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 3, figsize=(16, 3))

for ax, mu in zip(axs.flatten(), mus):
    y = stats.norm.cdf(x, loc=mu, scale=sigma)
    ax.plot(x, y, linewidth=2)
    ax.set_ylim(-0.1, 1.1)
    ax.set_xlim(-4, 4)
    ax.set_title("CDF from Normal({}, 0)".format(mu))

x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 3, figsize=(16, 3))

for ax, mu in zip(axs.flatten(), mus):
    y = stats.norm.pdf(x, loc=mu, scale=sigma)
    ax.plot(x, y, linewidth=2)
    ax.set_ylim(-0.1, 0.5)
    ax.set_xlim(-4, 4)
    ax.set_title("PDF from Normal({}, 0)".format(mu))

x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 3, figsize=(16, 3))

for ax, mu in zip(axs.flatten(), mus):
    normal_samples = rnd.normal(loc=mu, scale=sigma, size=2500)
    ax.hist(normal_samples, bins=50, normed=True)
    ax.set_xlim(-3, 3)
    ax.set_ylim(0, 0.6)
    ax.set_title("PMF from Normal({}, 0)".format(mu))
```

## Changing σ stretches and shrinks the distribution function

```python
mu = 0
sigmas = [0.5, 1, 2]

x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 3, figsize=(16, 3))

for ax, sigma in zip(axs.flatten(), sigmas):
    y = stats.norm.cdf(x, loc=mu, scale=sigma)
    ax.plot(x, y, linewidth=2)
    ax.set_ylim(-0.1, 1.1)
    ax.set_xlim(-4, 4)
    ax.set_title("CDF from Normal(0, {})".format(sigma))

x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 3, figsize=(16, 3))

for ax, sigma in zip(axs.flatten(), sigmas):
    y = stats.norm.pdf(x, loc=mu, scale=sigma)
    ax.plot(x, y, linewidth=2)
    ax.set_ylim(-0.1, 1.0)
    ax.set_xlim(-4, 4)
    ax.set_title("PDF from Normal(0, {})".format(sigma))

x = np.linspace(-4, 4, num=250)
fig, axs = plt.subplots(1, 3, figsize=(16, 3))

for ax, sigma in zip(axs.flatten(), sigmas):
    normal_samples = rnd.normal(loc=mu, scale=sigma, size=2500)
    ax.hist(normal_samples, bins=50, normed=True)
    ax.set_xlim(-3, 3)
    ax.set_ylim(0, 1.0)
    ax.set_title("PMF from Normal(0, {})".format(sigma))
```
