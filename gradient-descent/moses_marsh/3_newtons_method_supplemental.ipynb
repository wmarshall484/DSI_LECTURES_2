```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def newtons_method(f_prime, f_double_prime, initial_guess,
                   threshold, max_iter):
    x = float(initial_guess)
    iterations = 0
    x_history = [x]
    while f_prime(x) > threshold and iterations < max_iter:
        x = x - f_prime(x)/f_double_prime(x)
        x_history.append(x)
        iterations += 1
    print('newtons method took {} iterations'.format(iterations))
    return x_history

def gradient_descent(f, f_prime, initial_guess, threshold, 
                     max_iter, learning_rate):
    x = float(initial_guess)
    iterations = 0
    x_history = [x]
    while f_prime(x) > threshold and iterations < max_iter:
        x = x - learning_rate*f_prime(x)
        x_history.append(x)
        iterations += 1
    print('gradient descent took {} iterations'.format(iterations))
    return x_history

def f(x): return x**4+x**2+10
def f_prime(x): return 4*x**2 + 2*x
def f_double_prime(x): return 8*x + 2

initial_guess = 10
threshold = .001
max_iter = 1000
learning_rate = .01

newton_x_history = newtons_method(f_prime, 
                                   f_double_prime,
                                   initial_guess, 
                                   threshold,
                                   max_iter)
gradient_descent_x_history = gradient_descent(f, f_prime,
                                              initial_guess,
                                              threshold, 
                                              max_iter, 
                                              learning_rate)
```

```python
x_axis_values = np.linspace(-11, 11, 1000)
```

```python
true_f = np.apply_along_axis(f, 0, x_axis_values)
newton_f_vals = np.apply_along_axis(f, 0, newton_x_history)
plt.plot(x_axis_values, true_f, label='f')
plt.scatter(newton_x_history, newton_f_vals, c='k', label='newton steps')
plt.xlabel('x')
plt.ylabel(r'$f(x) = x^4 + x^2 + 10$')
plt.legend();
```

```python
gd_f_vals = np.apply_along_axis(f, 0, gradient_descent_x_history)
plt.plot(x_axis_values, true_f, label='f')
plt.scatter(gradient_descent_x_history, gd_f_vals, c='k', label='grad. desc. steps')
plt.xlabel('x')
plt.ylabel(r'$f(x) = x^4 + x^2 + 10$')
plt.legend();
```
