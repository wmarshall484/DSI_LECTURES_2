```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.decomposition import NMF
import matplotlib.pyplot as py
%matplotlib inline
```

```python
# generate cluster data
X = make_blobs(centers=2, n_features=4, center_box=(5,15),random_state=1)[0]
py.scatter(X[:,0], X[:,1])
```

```python
nmf_model = NMF(n_components=2)
U = nmf_model.fit_transform(X)
V = nmf_model.components_.T
```

```python
# compare predicted and actual values
print U.dot(V.T)[0:3,:]
print
print X[0:3,:]
```

```python
# label obs with clusters
cluster_labels = np.argmax(U, axis=1)
group_0 = X[cluster_labels==0]
group_1 = X[cluster_labels==1]
py.scatter(group_0[:,0],group_0[:,1],c='r')
py.scatter(group_1[:,0],group_1[:,1],c='b')
```

```python
py.scatter(group_0[:,1],group_0[:,2],c='r')
py.scatter(group_1[:,1],group_1[:,2],c='b')
```

```python
py.scatter(group_0[:,2],group_0[:,3],c='r')
py.scatter(group_1[:,2],group_1[:,3],c='b')
```
