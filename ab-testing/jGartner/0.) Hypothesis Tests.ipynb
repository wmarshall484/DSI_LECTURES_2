```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')
from mpltools import style
style.use('ggplot')
```

<h2>Exact Test</h2>
<p>What we have described is called an <a href="http://www.biostathandbook.com/exactgof.html">exact test</a>.  Exact tests are used when sample sizes are small, and we have a strong intuition as to what the underlying PDF/PMF should be.  Before we collect data, we set our significance level to be 0.05.  Now, we use the binomial PMF to determine what the p-value is for our observation of 4 out of 15 voters preferring our candidate.
</p>

```python
from scipy.stats import binom
fig, ax = plt.subplots(1, 1)
n, p = 15, 0.5
x = np.arange(0,16)
ax.bar(x, binom.pmf(x, n, p), 1, lw=1, edgecolor='black', label='Null Hypothesis')
ax.set_title("PMF of the Null Hypothesis")
ax.set_ylabel("Probability")
ax.set_xlabel("Number of people preffering your candidate")
plt.show()
```

<h2>Calcualte P-Value</h2>
<p>The p-value is the probability of the null hypothesis prodcucing an observation as extreme <b>or more so</b>.  Hence, the cumulative density function for at our observed value will tell use this probability.</p>

```python
for i in range(5):
    print("Probability for {} observed is {}".format(i, binom.pmf(i, n, p)))

print("Cumulative probability is: ", binom.cdf(4, n, p))
```

<h3>We fail to reject the null hypothesis</h3>
<p>We can see that our p-value is less extreme than our significance level, and therefore we fail to reject the null hypothesis.  Let's hop back to the slides to discuss some other types of tests.</p> </br> </br>
