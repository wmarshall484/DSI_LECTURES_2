```python
import pandas as pd
import numpy as np
import scipy.stats as scs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def clt(dist,sample_size,hist):
    means = []
    for i in range(0,1000):
        means.append(np.mean(dist.rvs(sample_size)))
    if hist:
        plt.hist(means,normed=1,alpha=0.2)
    x = np.linspace(dist.ppf(0.01),dist.ppf(0.99), 100)
    plt.plot(x, dist.pdf(x), 'k-', lw=2)
```

```python
clt(scs.uniform(),5,False)
```

```python
clt(scs.uniform(),10,True)
```

```python
clt(scs.gamma(1.5),1,False)
```

```python
clt(scs.gamma(1.5),30,True)
```

```python
clt(scs.gamma(1),1,False)
```

```python
clt(scs.gamma(1),30,True)
```
