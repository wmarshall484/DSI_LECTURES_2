```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.random.multivariate_normal([0,0], [[1,.8],[.8,1]], 150)
X = (X - X.mean(axis=0))/X.std(axis=0)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.xlabel('x1')
plt.ylabel('x2')
plt.axis('square')
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
V = pca.components_
v1 = V[0]
v2 = V[1]
```

```python
V
```

```python
v1.dot(v2)
```

```python
xx1 = np.linspace(-2, 2, 10)
xx2 = np.linspace(-0.7, 0.7, 10)
line1 = (v1[1]/v1[0])*xx1 
line2 = (v2[1]/v2[0])*xx2
```

```python
plt.scatter(X[:,0], X[:,1], color='k', s=10.8,alpha=0.7)
#lt.plot([0,v1[0]], [0,v1[1]], linewidth=5, color='b')
#lt.plot([0, v2[0]], [0,v2[1]], linewidth=5, color='g')
plt.plot(xx1, line1, linewidth=3, color='b')
plt.plot(xx2, line2, linewidth=3, color='g')
plt.axis('square')
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
plt.scatter(X[:,0], X[:,1], color='k', s=10.8,alpha=0.7)
plt.arrow(0, 0,v1[0],v1[1], linewidth=5, color='b')
plt.arrow(0, 0,v2[0],v2[1], linewidth=5, color='g')
plt.axis('square')
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
theta = np.pi/10
```

```python
rot = np.array([[np.cos(theta), np.sin(theta)],[-np.sin(theta), np.cos(theta)]])
```

```python
X_rot = X.dot(rot)
```

```python
line1 = np.array([[-3,0],[3,0]])
line2 = np.array([[0,-3],[0,3]])
line1_rot = line1.dot(rot)
line2_rot = line2.dot(rot)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.plot(line1_rot[:,0], line1_rot[:,1], color='g', linewidth=2, alpha=0.5)
plt.plot(line2_rot[:,0], line2_rot[:,1], color='g', linewidth=2, alpha=0.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.axis('square')
```

```python
plt.scatter(X_rot[:,0], X_rot[:,1])
plt.axhline(0,color='g')
plt.axvline(0, color='g')
plt.xlabel("x1'")
plt.ylabel("x2'")
plt.axis('square')
```
