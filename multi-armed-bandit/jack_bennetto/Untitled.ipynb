```python
import numpy as np
import scipy.stats as scs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rc('text', usetex=True)
plt.figure(figsize=(10,5))
x = np.linspace(0,1,100)
for (a,b,s) in [(1,1,"r-"), (2,1,"g-"), (3,1,"b-"), (2,2,"g--"), (3,2,"b--"), (3,3,"b:")]:
        beta = scs.beta(a,b)
        plt.plot(x, beta.pdf(x), s, label="({0},{1})".format(a,b))
plt.legend(title=r"($\alpha,\beta$)", loc="upper left")
plt.xlabel("p")
plt.ylabel("pdf")
plt.savefig("images/betadistributions.png")
#plt.show()
```
