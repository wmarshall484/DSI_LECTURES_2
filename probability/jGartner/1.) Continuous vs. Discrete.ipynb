```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')
from mpltools import style
style.use('ggplot')
```

<h2>Normal Distribution</h2>

```python
from scipy.stats import norm
x = np.linspace(1, 9, num=1000)
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x, 5, 1), 'r-', lw=5, alpha=0.6, label='norm pdf')
ax.set_title("Lifetime of a Fly")
ax.set_ylabel("Probability Density")
ax.set_xlabel("Days")
plt.show()
```

<h2>Normal Distribution w/ CDF</h2>

```python
from scipy.stats import norm
x = np.linspace(1, 9, num=1000)
fig, ax = plt.subplots(1, 1)
ax.plot(x, norm.pdf(x, 5, 1), 'r-', lw=3, alpha=0.6, label='norm pdf')
ax.plot(x, norm.cdf(x, 5, 1), 'g-', lw=2, alpha=0.6, label='norm cdf')
ax.set_title("Lifetime of a Fly")
ax.set_ylabel("Probability")
ax.set_xlabel("Days")
ax.legend(loc='upper left')
plt.show()
```

<h2>Bernoulli</h2>

```python
fig, ax = plt.subplots(1, 1)
n, p = 10, 0.5
x = np.arange(0,11)
ax.bar([0,1], [.6,.4], 1, fill=False, lw=1, edgecolor='red', label='bernoulli pmf')
ax.set_title("Unfair Coin Bernoulli")
ax.set_ylabel("Probability")
ax.set_ylim(0,1)
ax.set_xlim(-1,2)
ax.set_xlabel("Outcome")
plt.show()
```

<h2>Binomial</h2>

```python
from scipy.stats import binom
fig, ax = plt.subplots(1, 1)
n, p = 10, 0.4
x = np.arange(0,11)
ax.bar(x, binom.pmf(x, n, p), 1, fill=False, lw=1, edgecolor='red', label='binom pmf')
ax.set_title("Ten Coinflip - Unfair Coin PMF")
ax.set_ylabel("Probability")
ax.set_xlabel("Sum")
plt.show()
```

<h2>Geometric</h2>

```python
from scipy.stats import geom
fig, ax = plt.subplots(1, 1)
x = np.arange(1,11)
ax.bar(x, geom.pmf(x, .3), 1, fill=False, lw=1, edgecolor='black', label='geom pmf')
ax.set_title("Geometric PMF")
ax.set_ylabel("Probability")
ax.set_xlabel("N-th Trial")
plt.show()
```

<h2>Poisson</h2>

```python
from scipy.stats import poisson
fig, ax = plt.subplots(1, 1)
x = np.arange(0,11)
ax.plot(x, poisson.pmf(x, 1), 'r*--', label=r"$\lambda$ = 1")
ax.plot(x, poisson.pmf(x, 3), 'b<--', label=r"$\lambda$ = 3")
ax.plot(x, poisson.pmf(x, 5), 'g<--', label=r"$\lambda$ = 5")
ax.set_title("Poisson PMF")
ax.set_ylabel("Probability")
ax.set_xlabel("N Observed")
ax.legend(loc='upper right')
plt.show()
```

<h2>Continous</h2>

```python
from scipy.stats import uniform
fig, ax = plt.subplots(1, 1)
x = np.linspace(0.5, 3.5, 500)
ax.plot(x, uniform.pdf(x, loc=1, scale=2),'r-', lw=2, alpha=0.6, label='a=1, b=3')
ax.plot(x, uniform.pdf(x, loc=1.5, scale=1),'g-', lw=2, alpha=0.6, label='a=1.5, b=2.5')
ax.set_title("Uniform PDF")
ax.set_ylabel("Probability Density")
ax.set_xlabel("Value")
ax.legend(loc='upper right')
plt.show()
```

<h2>Normal</h2>

```python
from scipy.stats import norm
fig, ax = plt.subplots(1, 1)
x = np.linspace(-10, 10, 500)
ax.plot(x, norm.pdf(x, loc=1, scale=2),'r-', lw=2, alpha=0.6, label=r'$\mu=1, \sigma=2$')
ax.plot(x, norm.pdf(x, loc=0, scale=1),'g-', lw=2, alpha=0.6, label=r'$\mu=0, \sigma=1$')
ax.plot(x, norm.pdf(x, loc=-1, scale=3),'c-', lw=2, alpha=0.6, label=r'$\mu=-1, \sigma=3$')
ax.set_title("Gaussian PDF")
ax.set_ylabel("Probability Density")
ax.set_xlabel("Value")
ax.legend(loc='upper left')
plt.show()
```

<h2>Exponential</h2>

```python
from scipy.stats import expon
fig, ax = plt.subplots(1, 1)
x = np.linspace(0, 10, 500)
ax.plot(x, expon.pdf(x, scale=1),'r-', lw=2, alpha=0.6, label=r'$\mu=1, \sigma=2$')
ax.plot(x, expon.pdf(x, scale=2),'g-', lw=2, alpha=0.6, label=r'$\mu=0, \sigma=1$')
ax.plot(x, expon.pdf(x, scale=3),'c-', lw=2, alpha=0.6, label=r'$\mu=-1, \sigma=3$')
ax.set_title("Exponential PDF")
ax.set_ylabel("Probability Density")
ax.set_xlabel("Value")
ax.legend(loc='upper right')
plt.show()
```
