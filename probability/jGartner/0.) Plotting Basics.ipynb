```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import warnings, itertools
warnings.filterwarnings('ignore')
from mpltools import style
style.use('ggplot')
```

<h2>Histogram Mechanics</h2>
<p>Making histograms <em>can</em> be a useful, if you are careful about how you do things.  Let's start by generating some data with non-trivial structure to it.</p>

```python
entries = np.concatenate((np.zeros(2000), np.random.exponential(size=9000), [x for x in np.random.normal(4., 1., 2000) if x > 0]))
```

```python
# First pass - thoughtless plot
fig, ax = plt.subplots()
ax.hist(entries)
fig.show()
```

<h2>Boooo!</h2>
<p>The above plot has a number of issues.  First and foremost, all plots must have title's and axis labels.  Even if you're making something quickly, being able to come back a week later is entirely dependent upon good descriptions. Additionally, the grainularity of the bins is too coarse.</p>

```python
fig, ax = plt.subplots()
ax.hist(entries, bins=100)
ax.set_title("Example Histogram")
ax.set_ylabel("Count")
ax.set_xlabel("Independent Variable")
fig.show()
```

<h2>Making progress</h2>
<p>So now our plot is reasonble, but we have the not so uncommon problem that the initial bin dominates the plot, and the structure of the tail is very interesting.  One way of investigating is to use log axis.  In addition, the plot is showing the counts of variables.  Typically, we prefer to depict the frequencies for a particular bin, which is controlled by the "normed" parameter.</p>

```python
fig, ax = plt.subplots()
ax.hist(entries, bins=100, normed=1)
ax.set_title("Example Histogram")
ax.set_ylabel("Count")
ax.set_xlabel("Independent Variable")
ax.set_yscale("log", nonposy='clip')
fig.show()
```
