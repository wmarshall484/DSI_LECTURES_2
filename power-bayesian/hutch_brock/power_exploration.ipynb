```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import scipy.stats as sts
import math
import pylab

%matplotlib inline
```

```python
# Graphic showing how power relates to sample size, effect size, significance and std
def plot_power(n=100, alpha=.05, mu=0, mu2=1, std=1):
    crit_val = sts.norm.ppf(1 - alpha)

    # Curve 1
    variance = std ** 2 / math.sqrt(n)
    sigma = math.sqrt(variance)
    x = np.linspace(mu - 10 * sigma, mu + 10 * sigma, 100)
    plt.plot(x,mlab.normpdf(x, mu, sigma))

    # Curve 2, assuming the two curves have the same std
    plt.plot(x, mlab.normpdf(x, mu2, sigma))

    # Critical line
    plt.axvline(crit_val * sigma)
    x3 = np.linspace(crit_val * sigma, 3, 100)
    plt.fill_between(x3, mlab.normpdf(x3, mu2, sigma), color='red')
    plt.figure(figsize=(20,10))

    power = 1 - sts.norm(mu2, sigma).cdf(crit_val * sigma)
    print "Power of the test is the area highlighted in red."
    print "Significance: " + str(alpha)
    print "Effect Size: " + str(abs(mu - mu2))
    print "Sample size: " + str(n)
    print "Standard Deviation: " + str(std)
    print "Power: " + str(power)
    plt.show()
```

```python
plot_power()
```

```python
plot_power(alpha=.01)
```

```python
plot_power(n=1000)
```

```python
plot_power(mu2 = 0)
```

```python
plot_power(std=2)
```
