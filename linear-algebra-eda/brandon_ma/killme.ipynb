```python
A = np.load("data/sprite.npy")
A_bw = np.load("data/spritebw.npy")
```

```python
plt.imshow( A_bw, cmap="Greys" )
```

```python
def to_bw(img):
    r, g, b = A[:,:,0], A[:,:,1], A[:,:,2]
    return 0.2126*r + 0.7152*g + 0.0722*b
```

```python
A_bw = np.around( 1-to_bw(A)/255, decimals=3 )
```

```python
np.save("data/spritebw.npy", A_bw)
```

```python
plt.imshow(A_bw, cmap="Greys")
```
