# Pie Charts

Generally use bar chart instead.

Humans aren't good at comparing angles (except 180° and 90°).

Only use if

 * Showing proportions,
 * Halves (and maybe quarters matter), and especially if
 * Neighboring slices add up.
 

```python
sizes = {}
sizes[1994] = [47, 50, 244, 252, 49, 30]
sizes[1998] = [43, 47, 198, 298, 47, 36]
sizes[2002] = [47, 58, 190, 251, 55, 2]
sizes[2005] = [61, 46, 180, 222, 51, 54]
sizes[2009] = [93, 45, 194, 146, 68, 76]
sizes[2013] = [0, 56, 255, 193, 63, 64]
colors = ['yellow', 'blue', 'black', 'red', 'green', 'brown']
labels = ['FDP', 'CSU', 'CDU', 'SPD', 'Green', 'Left/PDS']
```

```python
fig = plt.figure(figsize=(17,11))
fig.suptitle("Seats in German Bundestag by Party", fontsize=16, fontweight='bold')
for nplot, year in ((231, 1994), (232, 1998),(233, 2002),(234, 2005), (235, 2009), (236, 2013)):
    ax = plt.subplot(nplot)
    ax.set_title(str(year), fontsize=14)
    ax.bar(range(1,7), sizes[year], tick_label=labels, color=colors, align='center')
plt.show()
```

This shows relative sizes of the parties but not which comprises a majority.

```python
fig = plt.figure(figsize=(17,11))
fig.suptitle("Seats in German Bundestag by Party", fontsize=16, fontweight='bold')
for nplot, year in ((231, 1994), (232, 1998),(233, 2002),(234, 2005), (235, 2009), (236, 2013)):
    ax = plt.subplot(nplot)
    ax.set_title(str(year), fontsize=14)
    ax.pie(sizes[year], colors=colors, labels=labels, startangle=270)
plt.show()
```

This shows which parties comprise a majority (assuming they are ordered by coalition preference) though it's hard to tell relative size.
